# Download Raw Data from Gdrive

```python
file_id = "1VfOg4cnR8mEG3mlSz7nknAQ7R_ewBMKx"

file = "payload_modeling_weekly_new_202404260930.xlsx"

gdown.download(f'https://drive.google.com/uc?export=download&id={file_id}', file, quiet=False)
```

```python
#file = r"C:\Users\kaziz\Downloads\Copy of 1. payload_modeling_weekly_new_202404260930.xlsx"

df = pd.read_excel(file, engine='openpyxl')

df.tail()
```

```python
df.isnull().sum()
```

```python
# cleaning kabupaten null

df1 = df[~df['id_kabupaten'].isna()]

df1.isnull().sum()
```

```python
# change data type for id_kabupaten

df1['id_kabupaten'] = df1['id_kabupaten'].astype('int64')

df1.tail()
```

```python
# create reference id_kabupaten, region_new, kabupaten

reff = df1[~df1['region_new'].isna()][['region_new','id_kabupaten','kabupaten']]
reff.tail()
```

```python
# merge data and reference to complete empty region_new

df1 = df1[['yearweek', 'id_kabupaten', 'kabupaten', 'count',
       'payload_gb', 'pgb', 'payload_per_user_daily_avg',
       'payload_per_user_week', 'prb_util_num', 'prb_util_denum', 'prb_util',
       'prb_cell_increase', 'prb_cell_high', 'cell_blank_coverage',
       'cell_decrease_coverage', 'count_tile_week', 'rhi_cell_good',
       'rhi_cell_fail', 'total_cell', 'total_cell_fail_cssr_data_service_4g',
       'total_cell_fail_data_retainability_success_rate_4g',
       'total_cell_fail_availability_4g', 'good_quality', 'good_coverage',
       'avg_download_throughput', 'avg_upload_throughput', 'avg_latency',
       'avg_packetloss', 'avg_jitter']].merge(reff, how = 'left', on=['id_kabupaten','kabupaten'])


df1.head()
```

```python
df1.isnull().sum()
```

```python
# remove kabupaten 'MAMUJU UTARA' [double id_kabupaten]

df1 = df1[df1['kabupaten']!='MAMUJU UTARA']

df1.head()
```

```python
df1.groupby('yearweek').count()
```

```python
# filter count == 7

df1 = df1[df1['count']==7]

df1.isnull().sum()
```

```python
# filter all value is null

df2 = df1[(~df1['total_cell'].isna())]
df2 = df2[(~df2['payload_per_user_week'].isna())]
df2 = df2[(~df2['cell_blank_coverage'].isna())]
df2 = df2[(~df2['good_quality'].isna())]
df2 = df2[(~df2['good_coverage'].isna())]
df2 = df2[(~df2['avg_download_throughput'].isna())]
df2 = df2[(~df2['prb_util_num'].isna())]


df2.isnull().sum()
```

```python
df2.groupby('yearweek').count()
```

```python
# save clean data to csv

df2.to_csv("payload_modeling_weekly_new_202404260930_clean.csv", index=False)
```
